```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#A categorical dataset to determine whether a user purchased a particular product
dataframe=pd.read_csv("/content/drive/MyDrive/AIML17/Social_Network_Ads.csv")
dataframe
```

```python
#count of people purchasing and not purchasing
v=dataframe['Purchased'].value_counts()
v
```

```python
v.plot(kind="bar",color=('b','g'))
plt.ylabel("Total no.of purchasing")
plt.xlabel("people Not purchased and purchased")
plt.show()
```

```python
x=dataframe.iloc[:,[2,3]].values
y=dataframe.iloc[:,4].values
```

```python
x
```

```python
y
```

```python
#split the data into train and test,put the test data to 30%
```

```python
from sklearn.model_selection import train_test_split as split
x_train,x_test,y_train,y_test=split(x,y,test_size=0.3,random_state=0)
```

```python
#Getting train and test data sizes
print(x_train.shape)
print(x_test.shape)
```

```python
#Logistic Regression is more efficient to train the model
from sklearn.linear_model import LogisticRegression as lcreg
model=lcreg()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
# the values in the train and test values are not matching with the expected values
#hence we need to normalize the inputs
# to normalize,we will use MinMaxScaler
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
plt.figure(figsize=(4,3))
ConfMatrix=confusion_matrix(y_test,y_pred)
sns.heatmap(ConfMatrix,annot=True,cmap="Greens",
            xticklabels=['Not purchased','purchased'],
            yticklabels=['Not purchased','purchased'])
plt.ylabel('Actual values')
plt.xlabel('Predicted values')
plt.title("Confusion matrix-Social_network_ads");
```
